```python
def get_coordinates(city_name):
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(city_name)
    if location:
        return location.latitude, location.longitude
    else:
       
        return None, None
```

```python
from datetime import datetime, timedelta
from meteostat import Point, Hourly, Daily


def calculate_dates(freq="hourly"):
    end_date = datetime.now()  
    if freq == "hourly":
        start_date = end_date - timedelta(hours=100) 
    elif freq == "daily":
        start_date = end_date - timedelta(days=100)  
    else:
        raise ValueError("Invalid frequency. Use 'hourly' or 'daily'.")
    return start_date, end_date


location = Point(23.0215374, 72.5800568)  


start_hourly, end_hourly = calculate_dates(freq="hourly")
hourly_data = Hourly(location, start_hourly, end_hourly).fetch()
print(f"Hourly Data Rows: {len(hourly_data)}")
print(hourly_data)

# For daily data
start_daily, end_daily = calculate_dates(freq="daily")
daily_data = Daily(location, start_daily, end_daily).fetch()
print(f"Daily Data Rows: {len(daily_data)}")
print(daily_data)
```
